## Importing necessary libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, f1_score, accuracy_score
```

## Loading the dataset

```python
cancer =  load_breast_cancer()
```

## Converting the data to a dataframe for easy understanding

## Storing the input variables to x

```python
# Dropping the 'Result' column from the DataFrame 'cancer_df' and storing the remaining columns in 'x'
x = cancer_df.drop(columns = 'Result' , axis = 1)
```

```python
x.head()
```

## Storing the output variable to y

```python
# Storing the 'Result' column from the DataFrame 'cancer_df' in the variable 'y
y = cancer_df['Result']
```

```python
y.head()
```

## Partitioning the dataset into training and testing subsets

```python
# 80% of the data will be used for training and 20% for testing
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size = 0.2,random_state= 2)
```

```python
#checking the no of rows of the training data
x_train.shape
```

```python
# checking the no of rows of testing data
x_test.shape
```

## Applying feature scaling using StandardScaler to normalize the dataset



```python
# Initializing the StandardScaler to standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
```

```python
# Applying standard scaling to the training data and converting it back to a DataFrame with original column names
x_train = pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
```

```python
# Applying standard scaling to the test data and converting it back to a DataFrame with original column names
x_test = pd.DataFrame(scaler.fit_transform(x_test),columns=x_test.columns)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_test.head()
```

## Algorithm Implimentation

### Logistic Regression

How it works


Logistic Regression models the probability that an instance belongs to a particular class (binary classification in this case: benign vs malignant) using a sigmoid function to convert the linear output into a probability. It assumes a linear relationship between the independent variables and the log-odds of the dependent variable.


Why suitable



It is interpretable and efficient for binary classification problems like this one. Logistic Regression works well when the features have a linear relationship with the output, making it a good starting point.

```python
# Creating an instance of the LogisticRegression class
lr_model = LogisticRegression()
```

```python
# Training the Logistic Regression model using the training data
lr_model.fit(x_train,y_train)
```

```python
# Making predictions on the test dataset
y_pred_lr = lr_model.predict(x_test)
```

```python
# Generate classification report
lr_report = classification_report(y_test, y_pred_lr, output_dict=True)
```

```python
lr_report_df = pd.DataFrame(lr_report).transpose()
```

```python
lr_report_df
```

```python
# Calculating the accuracy of the Logistic Regression model
accuracy_of_lr_model = accuracy_score(y_test,y_pred_lr)
```

```python
print(f"The accuracy of the logistic regression model is {accuracy_of_lr_model:.2f}")
```

##  Decision Tree Classifier


How it works

 This algorithm splits the data into subsets based on the value of features using conditions that maximize information gain or minimize impurity (Gini or entropy). It builds a tree where each internal node represents a decision on a feature, and each leaf represents a class label.

Why suitable

Decision Trees can handle both numerical and categorical data and are easy to visualize. They are non-linear models, so they can capture complex interactions between features. Since the breast cancer dataset has many features, this model can identify important ones and perform well.


```python
# Creating an instance of the Decisiom tree classifier class
dtc_model = DecisionTreeClassifier()
```

```python
# Training the Deciion tree classifier model using the training data
dtc_model.fit(x_train,y_train)
```

```python
# Making predictions on the test dataset
y_pred_dtc = dtc_model.predict(x_test)
```

```python
# Generate classification report
dtc_report = classification_report(y_test, y_pred_dtc, output_dict=True)
```

```python
dtc_report_df = pd.DataFrame(dtc_report).transpose()
```

```python
dtc_report_df
```

```python
# Calculating the accuracy of the Decision tree classifier  model
accuracy_of_dtc_model = accuracy_score(y_test,y_pred_dtc)
```

```python
print(f"The accuracy of the Decision Tree Classifier  model is {accuracy_of_dtc_model:.2f}")
```

##  Random Forest Classifier


How it works

Random Forest is an ensemble learning method that builds multiple decision trees and combines their predictions (via averaging for regression or majority vote for classification). Each tree is built using a random subset of features, reducing overfitting

Why suitable

Random Forest is robust to overfitting and handles high-dimensional datasets well, making it suitable for the breast cancer dataset with 30 features. It also gives insight into feature importance.

```python
# Creating an instance of the Random Forest Classifier class
rfc_model = RandomForestClassifier()
```

## Support Vector Machine


How it works

SVM tries to find a hyperplane that best separates the data points into two classes with the largest margin. It uses kernel tricks to handle non-linear relationships between features and the output.

Why suitable

SVM is effective in high-dimensional spaces and works well when the classes are separable.

```python
# Creating an instance of the Support Vector Machine class
svc_model = SVC()
```

##  k-Nearest Neighbors (k-NN)


How it works

 k-NN classifies a data point based on the majority class of its 'k' nearest neighbors in the feature space. Distance measures like Euclidean distance are used to find neighbors

Why suitable

k-NN is simple and effective, especially for small to medium-sized datasets. While it may be slower with large datasets, it can still work well here because the dataset isn't too large (569 instances), and the features are numeric, which is ideal for distance-based classification.

```python
# Creating an instance of the KNN class
knn_model =  KNeighborsClassifier()
```

#### Since the accuracy of the three algorithms are same we are using class 1 F1 score to compare the perfomance of the ml algorithms

```python
# Create a list of classification report dataframes for comparison across models
report_dfs = [lr_report_df,dtc_report_df,rfc_report_df,svc_report_df,knn_report_df]
```

```python
# Create a list of model names corresponding to the classification reports
model_names = ['Logistic Regression','Decision Tree','Random Forest','SVM','KNN']
```

```python
# Initialize an empty list to store the F1-scores for class 1 across different models
f1_scores_class_1 = []
```

```python
# Loop through each dataframe in the report_dfs list to extract the F1-scores for class 1
for df in report_dfs:
    f1_score = df.iloc[1, 2]
    f1_scores_class_1.append(f1_score)
```

```python
f1_scores_class_1
```

```python
# Create a dictionary that maps model names to their corresponding F1-scores for class 1
model_f1_scores = {model_names[i]: f1_scores_class_1[i] for i in range(len(model_names))}
```

```python
model_f1_scores
```

```python
df_f1_scores = pd.DataFrame(list(model_f1_scores.items()), columns=['Model', 'F1 Score'])
```

```python
df_f1_scores
```

## Finding the model with high perfomance

```python
# Find the key corresponding to the best f1 score
max_key = max(model_f1_scores, key=model_f1_scores.get)
print(f"Model with the highest F1-Score for Class 1: {max_key}")
```

```python
## Finding the model with low perfomance
```

```python
# Find the key corresponding to the worst f1 score
min_key = min(model_f1_scores, key=model_f1_scores.get)
print(f"Model with the lowest F1-Score for Class 1: {min_key}")
```
